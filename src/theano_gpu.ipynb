```python
import numpy as np
import sklearn
from sklearn import datasets
import theano
from theano import tensor
from theano.tensor.shared_randomstreams import RandomStreams
import time

from IPython.display import Image
from IPython.display import SVG
```

```python
# GPU uses float32, Uncomment the below line to set data type to float32
# theano.config.floatX = 'float32'
```

```python
# Testing on moon data
train_x, train_y = datasets.make_moons(10000, noise=0.1)
train_y_onehot = np.eye(2)[train_y]
# train_x = train_x.astype(np.float32)
# train_y = train_y.astype(np.float32)
hidden_dim = 1000
input_dim=2
output_dim=2
num_examples = len(train_x)
```

```python
# parameters for the model
lr = np.float32(0.01) # Learning rate
lmbda = np.float32(0.01) # regularization strength
```

```python
# initialize shared values on the gpu
rng = RandomStreams(seed=123)
x = theano.shared(train_x.astype('float32'), name='x')
y = theano.shared(train_y_onehot.astype('float32'), name='y')
# Glorot Uniform
w1 = theano.shared(np.random.uniform(
                        low = -np.sqrt(6./(input_dim + hidden_dim)),
                        high = np.sqrt(6./(input_dim + hidden_dim)),
                        size = (input_dim, hidden_dim)).astype('float32'), name='w1')
w2 = theano.shared(np.random.uniform(
                        low = -np.sqrt(6./(hidden_dim + output_dim)),
                        high = np.sqrt(6./(hidden_dim + output_dim)),
                        size = (hidden_dim, output_dim)).astype('float32'), name='w2')
b1 = theano.shared(np.zeros(hidden_dim).astype('float32'), name='b1')
b2 = theano.shared(np.zeros(output_dim).astype('float32'), name='b2')
```

```python
# Forward propagation
z1 = x.dot(w1) + b1
# Recitified Linear Units
a1 = tensor.maximum(0, z1)
z2 = a1.dot(w2) + b2
y_hat = tensor.nnet.softmax(z2)
```

```python
# Define the loss function
loss = 1./num_examples * lmbda/2 * (tensor.sum(tensor.sqr(w1)) + tensor.sum(tensor.sqr(w2)))
# Optimize the loss function
loss = tensor.nnet.categorical_crossentropy(y_hat, y).mean() + loss
# Returns a class prediction
prediction = tensor.argmax(y_hat, axis=1)
```

```python
# gradients
dw2 = tensor.grad(loss, w2)
dw1 = tensor.grad(loss, w1)
db1 = tensor.grad(loss, b1)
db2 = tensor.grad(loss, b2)
```

```python
# Remove the input values to use the same shared variable
forward_prop = theano.function([], y_hat)
calc_loss = theano.function([], loss)
predict = theano.function([], prediction)
```

```python
# Now assign the gradient update step
grad_step = theano.function(inputs=[],
                            updates=((w2, w2 - lr*dw2),
                                     (w1, w1 - lr*dw1),
                                     (b1, b1 - lr*db1),
                                     (b2, b2 - lr*db2)))
```

```python
def build_model(n_iter=2000, print_loss=False):
    np.random.seed(213)
    w1.set_value(np.random.uniform(
                        low = -np.sqrt(6./(input_dim + hidden_dim)),
                        high = np.sqrt(6./(input_dim + hidden_dim)),
                        size = (input_dim, hidden_dim)).astype('float32'))
    w2.set_value(np.random.uniform(
                        low = -np.sqrt(6./(output_dim + hidden_dim)),
                        high = np.sqrt(6./(output_dim + hidden_dim)),
                        size = (hidden_dim, output_dim)).astype('float32'))
    b1.set_value(np.zeros(hidden_dim).astype('float32'))
    b2.set_value(np.zeros(output_dim).astype('float32'))
    
    # DO gradient update for each batch
    for num in xrange(0, n_iter):
        # Perform gradient step
        grad_step()
        
        if print_loss and num%1000==0:
            print 'Loss after iteration',num,'is',calc_loss()
    print 'Final Loss:',calc_loss()
```

```python
%timeit grad_step()
```

```python
build_model(n_iter=10000,print_loss=True)
```
